```python
from matplotlib import pyplot as plt
from shapely.geometry.polygon import Polygon
from shapely.geometry import Point
from descartes import PolygonPatch

from cycsat.simulator import Agency
from cycsat.database import Site
from cycsat.geometry import place_features

%matplotlib inline
```

```python
hq = Agency('space agency')
site = hq.select(Site)
features = place_features(site)

fig = plt.figure(2, figsize=(5,5), dpi=90)

feat1 = features[0]
feat2 = features[1]

ax = fig.add_subplot(111)

canvas_patch1 = PolygonPatch(feat1)
canvas_patch2 = PolygonPatch(feat2)

ax.add_patch(canvas_patch1)
ax.add_patch(canvas_patch2)

ax.set_title('random site')

xrange = [0, 100000]
yrange = [0, 100000]

ax.set_xlim(xrange)
ax.set_ylim(yrange)
ax.set_aspect(1)
```
